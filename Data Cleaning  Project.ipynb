```python
data=data[[ 'name', 'country', 'start_year',   'end_year',    'Mat',       'Inns',         'NO',       'Runs',         'HS',
              'Ave',         'BF',         'SR',        100.0,         50.0,
                0.0,         '4s',         '6s']]
```

```python
data[["Mat"]]=data[["Mat"]].replace("[+*]","",regex=True)
```

```python
data[["BF","SR"]]=data[["BF","SR"]].replace("-",0)
```
